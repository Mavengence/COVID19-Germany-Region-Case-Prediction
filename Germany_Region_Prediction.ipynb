# (4) Aufgabe 1

Bei den oberen Landkreisen ("Fail at ..") gab es einfach zu wenig Daten vor bis inkl. 18. März. Ich konnte darauf nicht predicten. Das heißt alle unteren Landkreise konnten nicht predicted werden. Man hätte evtl noch eine Lineare Regression darauf laufen lassen können, aber ich glaube das wäre trotzdem nicht besonders aussagekräftig gewesen

```python
def RMSLE(predict, target):
    assert len(predict) == len(target)
    predict = np.array(predict)
    target = np.array(target)
    total = 0 
    for k in range(len(predict)):
        LPred= np.log1p(predict[k]+1)
        LTarg = np.log1p(target[k] + 1)
        if not (math.isnan(LPred)) and  not (math.isnan(LTarg)): 
            total = total + ((LPred-LTarg) **2)
        
    total = total / len(predict)        
    return np.sqrt(total)
```

```python
landkreis_dict = {landkreis: {'2020-3-20_PRED': None, 
                              '2020-3-20_TRUE': None,
                              '2020-3-27_PRED': None,
                              '2020-3-27_TRUE': None,
                              'rmsle': None,
                              'pred_uc': None,
                              'y': None,
                              'forecast': None} for landkreis in landkreise}
```

```python
for landkreis in landkreise:
    try:
        landkreis_df = df[df['Landkreis'] == landkreis].copy()

        y = landkreis_df['AnzahlFall']

        grouped_y = landkreis_df.groupby('Meldedatum').sum().copy()
        idx = pd.date_range(str(grouped_y.index[0].date()), '2020-03-27')
        grouped_y.index = pd.DatetimeIndex(grouped_y.index)
        grouped_y = grouped_y.reindex(idx, method='ffill')['AnzahlFall']
        
        try:
            y_train = grouped_y[grouped_y.index <= '2020-03-19'].rolling(window=2).mean()
            y_test = grouped_y[grouped_y.index > '2020-03-19']
            
            mod = sm.tsa.statespace.SARIMAX(y_train,
                                        order=(2, 1, 2),
                                        seasonal_order=(0, 2, 2, 2),
                                        enforce_stationarity=False,
                                        enforce_invertibility=False)
            results = mod.fit()

        except:
            y_train = grouped_y[grouped_y.index <= '2020-03-19'].rolling(window=2).mean()
            y_test = grouped_y[grouped_y.index > '2020-03-19']
            
            mod = sm.tsa.statespace.SARIMAX(y_train,
                                        order=(1, 1, 1),
                                        seasonal_order=(0, 1, 2, 2),
                                        enforce_stationarity=False,
                                        enforce_invertibility=False)

            results = mod.fit()            
        
        pred_uc = results.get_forecast(steps=8)
        y_forecast = pred_uc.predicted_mean        
        rmsle_loss = RMSLE(y_forecast, y_test)

        landkreis_dict[landkreis]['2020-3-20_PRED'] =  y_forecast['2020-03-20']
        landkreis_dict[landkreis]['2020-3-27_PRED'] =  y_forecast['2020-03-27']
        landkreis_dict[landkreis]['2020-3-20_TRUE'] =  y_test['2020-03-20']
        landkreis_dict[landkreis]['2020-3-27_TRUE'] =  y_test['2020-03-27']
        landkreis_dict[landkreis]['rmsle'] = rmsle_loss
        landkreis_dict[landkreis]['pred_uc'] = pred_uc
        landkreis_dict[landkreis]['y'] = grouped_y
        landkreis_dict[landkreis]['forecast'] = y_forecast
        
    except:
        print("Fail at", landkreis)
```

```python
def print_forecast_for_Landkreis(landkreis_dict, landkreis_name):
    y = landkreis_dict[landkreis_name]['y'] 
    y_forecast = landkreis_dict[landkreis_name]['forecast']

    pred_uc = landkreis_dict[landkreis]['pred_uc'] 
    pred_ci = pred_uc.conf_int()

    ax = y.plot(label='observed', figsize=(14, 4))
    y_forecast.plot(ax=ax, label='Forecast')
   
    ax.fill_between(pred_ci.index,
                    pred_ci.iloc[:, 0],
                    pred_ci.iloc[:, 1], color='k', alpha=.25)
    ax.set_xlabel('Date')
    ax.set_ylabel('Corona Fälle')
    ax.set_title(f'Landkreis: {landkreis_name}')
    plt.legend()
    plt.show()
```

## (5) Model Selektion für das Training

Weil ich **ARIMA** benutzt habe um für die Landkreise einzelne Vorhersagen zu machen, muss man eine Saisonalitäts Reihenfolge anlegen. Die habe ich vorher mit einem Test Landkreis (SK München), weil da recht viele Einträge bis inkl. 18. März waren verwendet, um eine Generalisierung für alle Landkreise zu machen.

```python
y_train = landkreis_dict['SK München']['y'].values
```

```python
p = d = q = s =range(0, 3)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], x[3]) for x in list(itertools.product(p, d, q, s))]
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
results_list = {}

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y_train,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=True)
            results = mod.fit()
            results_list[results.aic] = (param, param_seasonal)
            #print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
            
        except:
            continue
```

#### Vorhersage für 21. Juli

```python
landkreis_dict2 = {landkreis: {'2020-07-21_PRED': None, 
                               'pred_uc': None,
                               'y': None,
                               'forecast': None} for landkreis in landkreise}
```

```python
for landkreis in landkreis_dict2:
    try:
        landkreis_df = df[df['Landkreis'] == landkreis].copy()

        y = landkreis_df['AnzahlFall']

        grouped_y = landkreis_df.groupby('Meldedatum').sum().copy()

        idx = pd.date_range(str(grouped_y.index[0].date()), '2020-07-19')
        grouped_y.index = pd.DatetimeIndex(grouped_y.index)

        grouped_y = grouped_y.reindex(idx, method='ffill')['AnzahlFall']

        y_train = grouped_y.rolling(window=7).mean()
        
        mod = sm.tsa.statespace.SARIMAX(y_train,
                                    order=(2, 1, 1),
                                    seasonal_order=(0, 2, 2, 2),
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)

        results = mod.fit()
    
        pred_uc = results.get_forecast(steps=21)
        y_forecast = pred_uc.predicted_mean
        
        landkreis_dict2[landkreis]['2020-07-21_PRED'] =  y_forecast['2020-07-21']

        landkreis_dict2[landkreis]['pred_uc'] = pred_uc
        landkreis_dict2[landkreis]['y'] = grouped_y
        landkreis_dict2[landkreis]['forecast'] = y_forecast
    except:
        print("Fail at", landkreis)
```
